    Evaluation and Analysis

```python
dataTest = read_csv('Content_Training_Testing/content_user_test.csv')
no_recom = 20

def evaluate(name_algo, location_algo_output_file, eval_output_file):
    output_data = read_csv(location_algo_output_file)

    output_data["numberGamesUserHasInTest"] = 0
    output_data["numberRecommendationUserHas"] = 0
    output_data["ratio"] = 0

    for i, row in output_data.iterrows():
        userGames = dataTest[dataTest["user_id"] == row["user_id"]]["game"].tolist()
        output_data.at[i, 'numberGamesUserHasInTest'] = len(userGames)
        count = 0
        for j in range(1, no_recom+1):
            if row[j] in userGames:
                count += 1
        output_data.at[i, "numberRecommendationUserHas"] = count
        if len(userGames) != 0:
            output_data.at[i, "ratio"] = float(count/len(userGames))
    print(name_algo)
    print(output_data["ratio"].describe(include=[float]))
    print(output_data["numberRecommendationUserHas"].describe(include=[float]))
    print(output_data["numberGamesUserHasInTest"].describe(include=[float]))
    output_data.to_csv(eval_output_file,
                       columns=["user_id", "ratio", "numberRecommendationUserHas", "numberGamesUserHasInTest"], index=False)
```

```python
evaluate("Hybrid with content and collaborative", pathlib.Path('Test_Files/hybrid1.csv'), pathlib.Path('Content_Output/hybrid2_ratio.csv'))
```

```python
evaluate("Hybrid with content and collaborative 2", pathlib.Path('Test_Files/hybrid2.csv'), pathlib.Path('Content_Output/hybrid1_ratio.csv'))
```

Evaluation using rmse

```python
dataTest = read_csv('Content_Training_Testing/content_user_test.csv')
userSet = read_csv('steam_users_purchase_play.csv')
no_recom = 20

def evaluate_rmse(name_algo, location_algo_output_file, eval_output_file):
    output_data = read_csv(location_algo_output_file)

    output_data["numberGamesUserHasInTest"] = 0
    output_data["numberRecommendationUserHas"] = 0
    output_data["MSE"] = 0
    output_data["RMSE"] = 0

    for i, row in output_data.iterrows():
        userGamesinTest = dataTest[dataTest["user_id"] == row["user_id"]]["game"].tolist()
        totaluserGames =  userSet[userSet["user_id"]== row["user_id"]]["game"].tolist()
        userGamesinTestLen = len(userGamesinTest)
        totaluserGamesLen = len(totaluserGames)
        output_data.at[i, 'numberGamesUserHasInTest'] = len(userGamesinTest)
        count = 0
        for j in range(1, no_recom+1):
            if row[j] in userGamesinTest:
                count += 1
        output_data.at[i, "numberRecommendationUserHas"] = count
        sqr_diff = (userGamesinTestLen - count) ** 2
        mse = (sqr_diff) / totaluserGamesLen
        rmse = math.sqrt(mse)
        if len(userGamesinTest) != 0:
            output_data.at[i, "MSE"] = mse
            output_data.at[i, "RMSE"] = rmse
        else:
            output_data.at[i, "MSE"] = 0
            output_data.at[i, "RMSE"] = 0
    print(name_algo)
    print(output_data["MSE"].describe(include=[float]))
    print(output_data["numberRecommendationUserHas"].describe(include=[float]))
    print(output_data["numberGamesUserHasInTest"].describe(include=[float]))
    output_data.to_csv(eval_output_file,
                       columns=["user_id", "RMSE", "MSE", "numberRecommendationUserHas", "numberGamesUserHasInTest"], index=False)
```

```python
evaluate_rmse("Content with genre, publisher, developer", pathlib.Path('Test_Files/genre_pub_dev.csv'), pathlib.Path('Content_Output/genre_pub_dev_rmse.csv'))
```

```python
evaluate_rmse("Content with genre, tags, game-details", pathlib.Path('Test_Files/genre_tags_gamedetail.csv'), pathlib.Path('Content_Output/genre_tag_game_rmse.csv'))
```
